```python
# Import libraries/packages

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# Read in the data
df = pd.read_csv(r'C:\Users\Lisal\OneDrive\Desktop\Portfolio\Python\movies.csv')

```

```python
# Look at data

df.head()
```

```python
# Looking for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data types for columns

df.dtypes
```

```python
# Changing data type of columns

df['budget'] = df['budget'].astype('float64')
df['gross'] = df['gross'].astype('float64')
df['runtime'] = df['runtime'].astype('float64')
df['votes'] = df['votes'].astype('float64')



df.head()
```

```python
# Correct year column

df['yearcorrect'] = df['released'].astype(str).str[8:-16]

df.head()
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Look for duplicates

df.drop_duplicates()

df.head()
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earning')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# Look at correlation  
#pearson

df.corr(method='pearson')
```

```python
#kendall

df.corr(method='kendall')
```

```python
#spearman

df.corr(method='spearman')
```

```python
# High correlation between budget and gross
```

```python
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix of Movie Numeric Features')
plt.xlabel('Numeric Features')
plt.ylabel('Numeric Features')
plt.show()
```

```python
# Look at Company

df_numerized = df
for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
df_numerized.corr()
```

```python
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix of Movie Numeric Features')
plt.xlabel('Numeric Features')
plt.ylabel('Numeric Features')
plt.show()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_correlation = sorted_pairs[(sorted_pairs) > 0.5]
high_correlation
```
